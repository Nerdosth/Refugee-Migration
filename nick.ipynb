```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the three dataframes
refugee_df = pd.read_csv('Resources\demographic_ml_df.csv')
death_df = pd.read_csv('Resources\event_ml_df.csv')
```

```python
refugee_df
```

```python
refugee_summary = refugee_df.groupby(['year', 'country_origin']).agg({'total': 'sum'}).reset_index()
refugee_summary
```

```python
death_df['total_deaths'] = death_df['disaster_deaths'] + death_df['conflict_deaths']
death_df
```

```python
def filter_databases(country_name):
    # Filter the refugee data by country name
    filtered_refugee_df = refugee_summary[refugee_summary['country_origin'] == country_name]

    # Filter the death data by country name
    filtered_event_df = death_df[death_df['country_origin'] == country_name]

    fig, ax = plt.subplots()

    # Plot the affected data as a line chart with a red color
    ax.scatter(filtered_event_df['year'], filtered_event_df['total_deaths'], color='red', label='Deaths')

    # Plot the refugees data as a scatter plot with a blue color
    ax.scatter(filtered_refugee_df['year'], filtered_refugee_df['total'], color='blue', label='Total Refugees')

    ax.set_title(f'{country_name} \n Total Deaths vs Total Refugees per Year')
    ax.set_xlabel('Year')
    ax.set_ylabel('Number of People')
    ax.legend()
    plt.show()

    # Calculate the correlation
    correlation = filtered_refugee_df['total'].corr(filtered_event_df['total_deaths'])
    print(f"The correlation for {country_name} is {correlation:.2f}")


#Loop through each country and print out a chart. 
unique_countries = refugee_summary['country_origin'].unique()

for i, country in enumerate(unique_countries):
    filter_databases(country)
    if i == len(unique_countries) - 1:
        break
```
